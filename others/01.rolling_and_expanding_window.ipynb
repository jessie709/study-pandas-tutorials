# What is a Rolling or Expanding window?

> Source: [Rolling and Expanding Windows For Dummies](https://robotwealth.com/rolling-and-expanding-windows-for-dummies/)

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Random stock prices
data = [100,101,99,105,102,103,104,101,105,102,99,98,105,109,105,120,115,109,105,108]

# Create pandas DataFrame from list
df = pd.DataFrame(data, columns=['close'])

# Calculate a 5 period simple moving average
sma5 = df['close'].rolling(window=5).mean()

# Plot
plt.plot(df['close'], label='Stock Data', color='blue')
plt.plot(sma5, label='SMA', color='red')

plt.legend()
plt.show()
```

- Use rolling windows: What's the mean of the passt n values at this point in time?
- Use expanding windows: What's the mean of all the data available up to this point in time?

Expanding windows have a fixed lower bound. Only the upper bound of the window is rolled forward (the window gets bigger).

```python
# Random stock prices
data = [100, 101, 99, 105, 102, 103, 104, 101, 105,
        102, 99, 98, 105, 109, 105, 120, 115, 109, 105, 108]

# Create pandas DataFrame from list
df = pd.DataFrame(data, columns=['close'])

# Calculate full sample mean for reference
full_sample_mean = df['close'].mean()

# Calculate expanding window mean
expanding_mean = df['close'].expanding(min_periods=1).mean()

# Plot
plt.figure(figsize=(10, 5))
plt.plot(df['close'], label='Stock Data', color='blue')
plt.plot(expanding_mean, label='Expanding Mean', color='red')
plt.plot(sma5, label='SMA', color='yellow')
plt.axhline(full_sample_mean, label='Full Sample Mean',
            linestyle='--', color='green')

plt.legend()
plt.show()
```
