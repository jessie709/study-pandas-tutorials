```python
import pandas as pd
```

```python
df = pd.read_excel("./data/vacancy202111.xlsx")
df
```

```python
# 拆分"组织路径"字段
df[['level1', 'level2', 'level3']] = df['组织路径'].str.split('|', expand=True).applymap(lambda x: x.strip())
df
```

```python
grouped = df.groupby(["level2", 'level3'])
for name, grp in grouped:
    fileName = './output/空置清单-' + '-'.join(name) + '.xlsx'
    grp.to_excel(fileName)
print('write to file ok!')
```
